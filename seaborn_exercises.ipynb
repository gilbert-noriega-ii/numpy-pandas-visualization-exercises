```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pydataset import data

import seaborn as sns

iris = sns.load_dataset('iris')
iris
```

What does the distribution of petal lengths look like?

```python
sns.distplot(iris.petal_length)
plt.show()
```

Is there a correlation between petal length and petal width?

```python
sns.relplot(x = 'petal_width', y = 'petal_length', data = iris)
plt.show()
#there is a positive correlation between petal length and petal width
```

Would it be reasonable to predict species based on sepal width and sepal length?

```python
sns.relplot(x = 'sepal_width', y = 'sepal_length', hue = 'species', data = iris)
plt.show()
# it would be reasonable to predict the species based on sepal width and sepal length although it would not be 100% accurate due to the overlap between virginica and versicolor
```

```python
sns.relplot(x = 'petal_width', y = 'petal_length', hue = 'species', data = iris)
plt.show()

#The petal_width and petal_length seem to be a better fit to predicting species
#The data is grouped better and have clearer lines of seperation from species to species
```

Using the chipotle dataset from the previous exercise, create a bar chart that shows the 4 most popular items and the revenue produced by each.

```python
from env import host, password, username
def get_db_url(db, username=username, host=host, password=password):
    return f'mysql+pymysql://{username}:{password}@{host}/{db}'
chipotle = pd.read_sql('SELECT * FROM orders', get_db_url('chipotle'))
chipotle
```

```python
chipotle['item_price'] = chipotle.item_price.replace('$', '').astype(float)
top_four = chipotle.groupby('item_name').quantity.sum().sort_values(ascending=False).head(4)
top_four
```

```python
revenue = chipotle.groupby('item_name').item_price.agg('sum')
top_four_revenue = pd.merge(top_four, revenue, left_on = 'item_name', right_on = 'item_name', how = 'inner')
top_four_revenue
```

```python
plt.figure(figsize=(8, 4))
sns.barplot(x = top_four_revenue.index, y = 'item_price', data = top_four_revenue)
plt.title('Revenue for Top 4 Items at Chipotle')
plt.xlabel(' Item Name')
plt.ylabel('Revenue')
plt.show()
```

5) Load the sleepstudy data and read it's documentation. Use seaborn to create a line chart of all the individual subject's reaction times and a more prominant line showing the average change in reaction time.

```python
sleep = data('sleepstudy')
data('sleepstudy', show_doc = True)
```

```python
#to get the overall reaction time, we will group the data 
#by days and get the mean from all subjects on those days
avg_reaction_time = pd.DataFrame(sleep.groupby('Days').Reaction.agg('mean'))
```

```python
#enlarging the size of the graph to better analyze the data
plt.figure(figsize=(10, 10))
#creating a line plot for the sleep data.
#the palette will change the lines to pastel so they aren't so prominent
sns.lineplot(x ='Days', y = 'Reaction', hue = 'Subject', data = sleep, palette = 'pastel')
#adding the average line onto the plot with the line bolder and a different shade
#give it a black color so it is much bolder than the rest then label it for the legend
plt.plot(avg_reaction_time.index, avg_reaction_time.Reaction, linestyle = 'dashdot', linewidth = 3, c = 'black', label = 'average')
#give the graph and y-axis a title
plt.title('Sleep Deprivation Study')
plt.ylabel('Reaction Time(ms)')
#show the legend
plt.legend()
plt.show()
```
