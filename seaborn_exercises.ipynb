Use the iris database to answer the following quesitons:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pydataset import data

import seaborn as sns

iris = sns.load_dataset('iris')
iris
```

What does the distribution of petal lengths look like?

```python
sns.distplot(iris.petal_length)
plt.show()
```

Is there a correlation between petal length and petal width?

```python
sns.relplot(x = 'petal_width', y = 'petal_length', data = iris)
plt.show()
#there is a positive correlation between petal length and petal width
```

Would it be reasonable to predict species based on sepal width and sepal length?

```python
sns.relplot(x = 'sepal_width', y = 'sepal_length', hue = 'species', data = iris)
plt.show()
# it would be reasonable to predict the species based on sepal width and sepal length although it would not be 100% accurate due to the overlap between virginica and versicolor
```

Which features would be best used to predict species?

```python
sns.relplot(x = 'petal_width', y = 'petal_length', hue = 'species', data = iris)
plt.show()

#The petal_width and petal_length seem to be a better fit to predicting species
#The data is grouped better and have clearer lines of seperation from species to species
```

1) Using the lesson as an example, use seaborn's load_dataset function to load the anscombe data set. Use pandas to group the data by the dataset column, and calculate summary statistics for each dataset. What do you notice?


```python
anscombe = sns.load_dataset('anscombe')

x = anscombe.groupby('dataset').x.describe()
y = anscombe.groupby('dataset').y.describe()
xy = pd.merge(x, y, left_on = 'dataset', right_on = 'dataset', how = 'inner')
xy

# anscombe_data = anscombe.groupby('dataset').agg(['max', 'min', 'mean','std'])
# anscombe_data

#the mean, std and number of x and y are very similar across all of the dataset
```

2) Load the InsectSprays dataset and read it's documentation.

```python
insect_sprays = data('InsectSprays')
#data('InsectSprays', show_doc=True)
insect_sprays
```

Create a boxplot that shows the effectiveness of the different insect sprays.

```python
sns.boxplot(data = insect_sprays, y = 'count', x = 'spray')
plt.title('Effectiveness of Different Insect Sprays')
plt.xlabel('Different Spray Types')
plt.ylabel('Number of Insects Left')
plt.show()
```
