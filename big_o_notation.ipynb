Curves to graph
y=0x+1 and label the curve "O(1)"
y=log(x) and label the curve "O(log n)"
y=x and label the curve "O(n)"
y=x*log(x) and label it "O(n log n)"
y=x^2 and label it "O(n^2)"
y=2^x and label it "O(2^n)"
y=n! and label it "O(n!)"
y=n^n and label it "O(n^n)"

```python
#importing matplot library
import matplotlib.pyplot as plt
#import math functions
import math

#in order to use the y equations you have to set a range for x
x = range(1,9)
y1 = [0 * n + 1 for n in x]
y2 = [math.log(n) for n in x]
y3 = [n for n in x]
y4 = [n * math.log(n) for n in x]

#plotting y1-y4 on a plot
#color line to distinguish different functions on a chart
#give it a label so it will produce on a legend
plt.plot(x,y1, c = 'firebrick', label = 'O(1)')
plt.plot(x,y2, c = 'black', label = '$O(log n)$')
plt.plot(x,y3, c = 'darkorange', label = 'O(n)')
plt.plot(x,y4, c = 'gold', label = '$O(n log n)$')

#give a title to the plot
plt.title('Big O Nation')
#label the x-axis
plt.xlabel('Elements')
#label the y-axis
plt.ylabel('Operations')
#create a legend
plt.legend()
#produce credit in corner
plt.text(10,1,'By: Gilbert Noriega')
#get rid of syntax on top of chart/show chart
plt.show()
```

```python
x1 = range(1,7)
y5 = [n ** 2 for n in x1]
y6 = [2 ** n for n in x1]

plt.plot(x1,y5, c = 'green', label = '$O(n^2$)')
plt.plot(x1,y6, c = 'blue', label = '$O(2^n)$')

plt.title('Big O Nation')
plt.xlabel('Elements')
plt.ylabel('Operations')
plt.legend()
plt.show()
```

```python
x2 = range(0,4)
y7 = [math.factorial(n) for n in x2]
y8 = [n ** n for n  in x2]

plt.plot(x2,y7, c = 'darkviolet', label = '$O(n!)$')
plt.plot(x2,y8, c = 'slategrey', label = '$O(n^n)$')

plt.title('Big O Nation')
plt.xlabel('Elements')
plt.ylabel('Operations')
plt.legend()
plt.show()
```
