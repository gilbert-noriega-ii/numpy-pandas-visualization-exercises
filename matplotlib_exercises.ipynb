```python
import matplotlib.pyplot as plt

x = range(-10,11)
y = [n**2 - n + 2 for n in x]

plt.plot(x,y)
plt.show()
```

Create and label 4 separate charts for the following equations (choose a range for x that makes sense):

You can use functions from the math module to help implement some of the equations above.

```python
x = range(0,10)

plt.plot(x)
plt.title('X')
plt.show
```

```python
import math
x = range(0,100)
y = [math.sqrt(n) for n in x]

plt.plot(x,y)
plt.title('Squareroot of X')
plt.show()
```

```python
x = range(0,10)
y = [n**3 for n in x]

plt.plot(x,y)
plt.title('X Cubed')
plt.show()
```

```python
x = range(0,10)
y = [2**n for n in x]

plt.plot(x,y)
plt.title('2^x')
plt.show()
```

Combine the figures you created in the last step into one large figure with 4 subplots.

```python
x = range(0,10)
x1 = range(0,100)
y1 = [math.sqrt(n) for n in x1]
y2 = [n**3 for n in x]
y3 = [2**n for n in x]

plt.figure(figsize=(8, 8))

plt.subplot(2,2,1)
plt.plot(x1)
plt.title('X')

plt.subplot(2,2,2)
plt.plot(x1,y1)
plt.title('Squareroot of X')

plt.subplot(2,2,3)
plt.plot(x,y2)
plt.title('X Cubed')

plt.subplot(2,2,4)
plt.plot(x,y3)
plt.title('2^x')

plt.show()
```

Comine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points. Be sure to include a legend.

```python
x = range(0,5)
y1 = [math.sqrt(n) for n in x]
y2 = [n**3 for n in x]
y3 = [2**n for n in x]

plt.figure(figsize=(8, 8))

plt.plot(x, c = 'orange', label = 'X')
plt.plot(x,y1, c = 'red', label = 'Squareroot of X')
plt.plot(x,y2, c = 'green', label = 'X Cubed')
plt.plot(x,y3, c = 'royalblue', label = '2^x')
plt.legend()
plt.title('Variations of X')
plt.show()
```
