1) Copy the code from the lesson to create a dataframe full of student grades.

```python
import pandas as pd
import numpy as np

np.random.seed(123)

students = ['Sally', 'Jane', 'Suzie', 'Billy', 'Ada', 'John', 'Thomas',
            'Marie', 'Albert', 'Richard', 'Isaac', 'Alan']

# randomly generate scores for each student for each subject
# note that all the values need to have the same length here
math_grades = np.random.randint(low=60, high=100, size=len(students))
english_grades = np.random.randint(low=60, high=100, size=len(students))
reading_grades = np.random.randint(low=60, high=100, size=len(students))

df = pd.DataFrame({'name': students,
                   'math': math_grades,
                   'english': english_grades,
                   'reading': reading_grades})
df
```

1a) Create a column named passing_english that indicates whether each student has a passing grade in reading.

```python
df['passing_english'] = df.english >= 70 
#"series['new_column_name']" creates a new column 
#in that series. the column will be filled by
#the data proceeding the equal sign
#in this case it will be filled with T for 
#all english grades 70 and F for the remainder
df
```

1b) Sort the english grades by the passing_english column. How are duplicates handled?

```python
df.sort_values('passing_english')
#"series.sort_values('condition')" will sort 
#the series by the condition
#in this instance the values will be sorted(ascending)
#based off of T/F values

#The duplicates are sorted by the index value.
```

1c) Sort the english grades first by passing_english and then by student name. All the students that are failing english should be first, and within the students that are failing english they should be ordered alphabetically. The same should be true for the students passing english. (Hint: you can pass a list to the .sort_values method)

```python
df.sort_values(['passing_english', 'name'])
#the values in the dataframe can be sorted by a list
#the values will be sorted by T/F from 
#'passing_english' then by 'name'
```

1d) Sort the english grades first by passing_english, and then by the actual english grade, similar to how we did in the last step.

```python
df.sort_values(['passing_english', 'english'])
#just like above, the values from the database will
#be sorted by the T/F values from 'passing_english'
#then by the grades from 'english'
```

1e) Calculate each students overall grade and add it as a column on the dataframe. The overall grade is the average of the math, english, and reading grades.

```python
df['overall_grade'] = (df.math + df.english + df.reading) / 3
df
#to find the overall grade you first need to add the grades from 
#'math', 'english', and 'reading' then divide it by 3
#this value will be stored in the new column created
#called 'overall_grade'
```

3g) How many animals are hoppers that are above the median speed? 
What percentage is this?

```python
hoppers = mammals[mammals.hoppers]
#create a subdataframe called 'hoppers' that will
#access the mammals dataframe and retrieve mammals
#under the 'hoppers' column 
number_of_hoppers = len(hoppers[hoppers.speed > mammals.speed.median()])
#create a variable callled number_of_hoppers that
#finds the total amount(len) of True values under the 'speed'
#column from the hoppers subdataframe that is 
#greater than the median value under the 'speed'
#column from the mammals dataframe

percentage = number_of_hoppers / len(mammals) * 100
#create a variable called percentage that takes
#in the 'number_of_hoppers' and divides it by the total
#amount of mammals(rows) then multiplies by 100
#to find the percentage value
percentage
```
